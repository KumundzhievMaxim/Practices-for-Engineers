Create 3 networks with networkx: a ring of 4, 5 and 6 nodes!

Select two nodes from each of the networks, and connect them with each other. In this way a clique of rings emerges.

Rewire this network by connecting two unconnected nodes with the highest clustering coefficients, and remove one edge, that connects one of the node pairs with the smallest clustering coefficients! (The sum of the clustering coefficients of the ending nodes of the link should be minimal.)

Repeat this rewiring step several times, and track changes of the degree distribution! Try to make a figure, which shows the change in the distribution and interpret the result! (Recall, what are the single valued descriptors of a distribution?)



```python
def ring_network(num_nodes):
    "Create a ring network with first- and second neighbor connections"
    ring = nx.Graph()
    for i in range(num_nodes):
        ring.add_edge(i, (i+1)%num_nodes)
        ring.add_edge(i, (i+2)%num_nodes)
    return ring

graphs = []
for num_nodes in range(4, 7):
    graphs.append(ring_network(num_nodes))
    
graphs    
#[(nx.draw_circular((graph), with_labels = True)) for graph in graphs]

[(nx.draw_circular((graphs[i]), with_labels = True), plt.show()) for i in range(len(graphs))]    
```
