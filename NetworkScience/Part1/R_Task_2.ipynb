```python
#Final Solution
def gen_grid_graph(i):
    """
    Generate a grid graph with about 2**i nodes.
    """
    n = 2**i
    sn = int(n**(1/2))
    # Redefine amount of nodes:
    print(n, sn)
    return nx.grid_2d_graph(sn,sn)
```

```python
#The number of nodes should be between 95 and 119.
#2 ** i where i from 6 upto 7
plt.figure(figsize=(10, 7))

G = gen_grid_graph(7)

nx.draw_networkx(G)
plt.show()
```

```python
from networkx.algorithms.planarity import check_planarity
#We are checking graph for planarity:
#Planarity - graph doesnt have interseptions

print(check_planarity(G))

plt.figure(figsize=(10, 7))
nx.draw_planar(G, with_labels = True, alpha=0.8)
plt.show()
```

 - To find minimal number of edge crossings and in general to understand do we have edge crossing, we can check our graph for planarity property (Reversly).

 - Planar graph - is a graph that can be embedded in the plane, i.e., it can be drawn on the plane in such a way that its edges intersect only at their endpoints. In other words, it can be drawn in such a way that no edges cross each other. 

 - So, we can check if our graph is planar, it means we can plot it in proper way. For that use check_planarity and draw_planar
 
 - In fact, the network is always able to guarantee a certain connectivity until the number of nodes removed are the biggest part of it
 
 - The sum of the degrees of the vertices of any graph is equal to twice the number of its edges. That is, the greater the number of vertices, the more (2 times more edges)
