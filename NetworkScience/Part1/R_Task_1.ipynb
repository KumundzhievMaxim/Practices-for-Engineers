```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
```

```python
G = nx.MultiDiGraph()

G.add_weighted_edges_from([(96, 97, 1.3),
                          (96, 98, 5.6),
                          (97, 98, 1.6),
                          (97, 99, 6.4),
                          (98, 96, 10.3),
                          (99, 96, 32.7),
                          (99, 97, 11.1),
                          (99, 98, 3.4)])

#Visualization
# fixing the size of the figure 
plt.figure(figsize =(5, 5)) 
  
nx.draw_networkx(G, with_labels = True)
nx.draw_networkx_edge_labels(G, pos=nx.spectral_layout(G));
plt.show()
```
