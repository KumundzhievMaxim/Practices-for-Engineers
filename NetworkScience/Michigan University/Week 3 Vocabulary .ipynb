### Degree and Closeness Centrality

#### Different ways of thinking about 'Importance'
 - Degree of nodes
 - Average proximity tp other nodes
 - Fraction of shortest pathes that pass through node
 - <b>Network Centratility</b> - measures identify the most important nodes in a network 
 - <b>Closeness Centratility</b> - important nodes are close to other nodes

- <b>Network Centratility</b> - measures identify the most important nodes in a network 
 - Degree Centarlity - important nodes have many connections;
     - The most basic measure of centrality is number of neighbours
  - Degree Centarlity (Undirected) $\frac{d_v}{|N| - 1}$
  - Degree IN Centarlity (Directed) $\frac{(d_v)^(in)}{|N| - 1}$, N - set of nodes, $(d_v)^(in)$ - the in-degree of v
  - Degree OUT Centarlity (Directed) $\frac{(d_v)^(out)}{|N| - 1}$, N - set of nodes, $(d_v)^(out)$ - the out-degree of v
- <b>Closeness Centratility</b> - important nodes are close to other nodes
 - Closeness Centratility - $\frac{|N| - 1}{Sum d(v,u)}$


### Example with Karate Club (Calculate degree centrality --> Undirected)

```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib notebook

G = nx.karate_club_graph()
G = nx.convert_node_labels_to_integers(G, first_label = 1)

#Calculate Degree centrality Undirected
degCent = nx.degree_centrality(G)

# #Calculate Degree in centrality Directed graph
# indegCent = nx.in_degree_centrality(G)

# #Calculate Degree out centrality Directed graph
# indegCent = nx.out_degree_centrality(G)

#Calculate Closeness Centrality Undirected
closeCentr = nx.closeness_centrality(G)

# #Calculate Closeness Centrality Directed
# closeCentr = nx.closeness_centrality(G, normalized = False)

nx.draw_networkx(G)
print('Calculate Degree centrality Undirected', degCent, '\n')
print('Closeness Centrality Undirected', closeCentr, '\n')
```
