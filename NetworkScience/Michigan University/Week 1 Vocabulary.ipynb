 - Network (Graph) - a representation of connections among a set of items
 - Items are called <b>Nodes</b>
 - Connections between items are called <b>edges</b> or <b>links</b> or <b>ties</b>
  - There are next types of networks:<br>
       1) <b>Undirected Network</b><br>
       2) <b>Directed Network</b><br>
       3) <b>Weighted Network</b><br>
       - Weighted networks - network where edges are assigned with a (typically numerical) weight;<br>
       4) <b>Signed Network</b><br>
       5) <b>MultiGraph Network</b><br>
       - MultiGraph Networks - A pair of nodes can have different types of relationships simultaneously;<br>

The Networks can have different types of attributes. <br>
As it was mentioned before, they can have such attributes as <b>Weightes</b>, <b>Signs</b>, <br>
but also another ones as <b>relation = 'Friend' </b>, <b>relation = 'Enemy' </b> and so on <br>

### NetworkX

 - How to create a Network in NetworkX?

```python
import matplotlib as plt
import networkx as nx
```

### Undirected Network (Graph)

```python
#Imports

#Define a Network (Graph) 
#--> Undirected
my_network = nx.Graph()

#Add edges A and B
my_network.add_edge('A', 'B')
my_network.add_edge('B', 'C')

#Plot the network (graph)

nx.draw(my_network);
```

### Directed Network (Graph)

```python
#Define a Network (Graph) 
#--> Directed
my_network_Directed = nx.DiGraph()

#Add edges A and B
my_network_Directed.add_edge('A', 'B')
my_network_Directed.add_edge('A', 'C')
my_network_Directed.add_edge('B', 'C')

#Plot the network (graph)

nx.draw(my_network_Directed);
```

### Weighted Network (Graph)

```python
#Define a Network
#First way of implementation using class nx.Graph()

my_network_weighted = nx.Graph()

my_network_weighted.add_edge('A', 'B', weight = 6)
my_network_weighted.add_edge('C', 'E', weight = 25)
my_network_weighted.add_edge('B', 'C', weight = 13)


#Plot the network (graph)
nx.draw(my_network_weighted);
```

### Signed Network (graph)


```python
G = nx.Graph()

G.add_edge('A', 'B', sign = '+')
G.add_edge('B', 'C', sign = '-')
G.add_edge('C', 'E', sign = '-')
G.add_edge('E', 'B', sign = '+')

#nx.draw(G)
nx.draw_networkx(G)
```

### MultiGraph Network (graph)

```python
G = nx.MultiGraph()

G.add_edge('A', 'B', relation = 'Friend')
G.add_edge('A', 'B', relation = 'Neighbour')

#nx.draw(G)   
nx.draw_networkx(G)
```

### Node and Edge Attributes (How to get the info about attributies)

```python
G = nx.DiGraph()
G.add_edge('A', 'B', weight = 6, relation = 'family')
G.add_edge('B', 'C', weight = 13, relation = 'friend')
G.add_edge('C', 'F', weight = 21, relation = 'coworker')
G.add_edge('C', 'E', weight = 25, relation = 'neighbour')
G.add_edge('F', 'G', weight = 9, relation = 'friend')
G.add_edge('E', 'D', weight = 2, relation = 'coworker')
G.add_edge('E', 'H', weight = 9, relation = 'coworker')
G.add_edge('E', 'I', weight = 10, relation = 'coworker')
G.add_edge('E', 'J', weight = 15, relation = 'family')
G.add_edge('I', 'J', weight = 3, relation = 'coworker')

# pos : dictionary
#    A dictionary with nodes as keys and positions as values.
#    Positions should be sequences of length 2.

pos = nx.spring_layout(G)

nx.draw_networkx(G, pos);
```

```python
#List of edges --> G.edges()
print('List of edges', '\n', G.edges(), '\n')

#List of edges with all atribbutes --> G.edges(data = True)
print('List of edges with all atribbutes', '\n', G.edges(data = True), '\n')

#List of edges with particular atribbute --> G.edges(data = 'relation')
print('List of edges with particular atribbute', '\n', G.edges(data = 'relation'), '\n')

#Access atribbutes of a specific edge --> G['A']['B']
print('Access atribbutes of a specific edge', '\n', G['A']['B'], '\n')

#Access  particular atribbutes of a specific edge --> G['A']['B']['relation'] / G['A']['B']['weight']
print('Access  particular atribbutes of a specific edge', '\n', G['A']['B']['relation'], '\n')
```

### Bipartite Graphs (двудольный граф) <br>
Двудо́льный граф или бигра́ф — это математический термин теории графов, обозначающий граф, множество вершин которого можно разбить на две части таким образом, что каждое ребро графа соединяет какую-то вершину из одной части с какой-то вершиной другой части, то есть не существует ребра, соединяющего две вершины из одной и той же части.

<img src="https://upload.wikimedia.org/wikipedia/commons/e/e8/Simple-bipartite-graph.svg?uselang=ru"
     alt="Markdown Monster icon"
     style="float: left; margin-right: 10px;" />

```python
#Define Bipartite Network (graph) 
from networkx.algorithms import bipartite

#Defined Graph
B = nx.Graph()

#Defined nodes in two different "sets" U and V
B.add_nodes_from(['A', 'B', 'C', 'E'], bipartite = 0)
B.add_nodes_from([1,3,4], bipartite = 1)

#Defined edges between two different "sets" U and V
B.add_edges_from([('A', 1), ('B', 1), ('C', 1), ('C', 3), ('E', 3), ('E', 4)])


#Draw Bipartite Graph
#nx.draw(B, with_labels = True)
nx.draw_networkx(B)
```

### Bipartite Graphs --> Projected Graphs<br>
    L - Bipartite graph projection - Network of nodes in group L, where a pair of nodes is connected if they have a common neighbour in R in the bipartite group.  

```python
#Define Projected Graph
B = nx.Graph()

B.add_edges_from([('A', 1), ('B', 1), ('C', 1), ('D', 1), ('H', 1), ('B', 2), ('C', 2),
                 ('D', 2), ('E', 2), ('G', 2), ('E', 3), ('F', 3), ('H', 3), ('J', 3),
                 ('E', 4), ('I', 4), ('J', 4)])

X = set(['A','B','C','D','E','F','G','H','I','J'])
P = bipartite.projected_graph(B, X)
#bipartite.weighted_projected_graph(B, X) --> for weighted case when edges are weighted

#It is a Projected Graph 
nx.draw_circular(P, with_labels = True)
```

### Adjacency List

`G_adjlist.txt` is the adjaceny list representation of G1.

It can be read as follows:
* `0 1 2 3 5` $\rightarrow$ node `0` is adjacent to nodes `1, 2, 3, 5`
* `1 3 6` $\rightarrow$ node `1` is (also) adjacent to nodes `3, 6`
* `2` $\rightarrow$ node `2` is (also) adjacent to no new nodes
* `3 4` $\rightarrow$ node `3` is (also) adjacent to node `4` 

and so on. Note that adjacencies are only accounted for once (e.g. node `2` is adjacent to node `0`, but node `0` is not listed in node `2`'s row, because that edge has already been accounted for in node `0`'s row).
