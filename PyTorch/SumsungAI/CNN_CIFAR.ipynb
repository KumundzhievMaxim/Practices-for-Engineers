```python
import torch
import numpy 
import random
import torchvision.datasets as data
```

```python
#Download Data
```

```python
train = data.CIFAR10('./', download=True)
test = data.CIFAR10('./', download=True)
```

```python
X_train = torch.FloatTensor(train.data)
X_test = torch.FloatTensor(test.data)

y_train = torch.LongTensor(train.targets)
y_test = torch.LongTensor(test.targets)
```

```python
print(
    X_train.shape,
    X_test.shape,
    y_train.shape,
    y_test.shape,
)
```

```python
#Normalize Images, cause now we have from 0 to 255. 
#We would like to have from 0 to 1
X_train /= 255.
X_test /= 255.
```

```python
#Plot few Images
import matplotlib.pyplot as plt
for i in range(5):
    plt.imshow(X_train[i])
    plt.title(str(y_train[0]))
    plt.show()
```
