Cross-modal Deep Variational Hand Pose Estimation


---


It is a novel approach to estimate the full 3D joint position of the hand from soley
monocular RGB images. The approach learns a manifold of valid hand poses with the help
of a Variational Autoencoder inwhich an RGB image can be projected into to retrieve the
displayed hand configuration, reaching state-of-the-art performance on the tested data
sets.

SPURR, Adrian, et al. Cross-modal deep variational hand pose estimation.
In: Proceedings of the IEEE Conference on Computer Vision and Pattern Recognition. 2018. p. 89-98.

http://openaccess.thecvf.com/content_cvpr_2018/CameraReady/3284.pdf
https://ait.ethz.ch/projects/2018/vae_hands/
